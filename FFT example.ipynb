```python
import numpy as np
import matplotlib.pyplot as plt

wzero = float(input("Enter the frequency: "))
alpha = float(input("Enter the alpha: "))
Npts = 10000
total_dura = 100
dt = 1/Npts*total_dura

t = np.arange(0,Npts)*dt
f = np.exp(-alpha*t)*np.cos(wzero*t)

plt.subplot(2,1,1)
plt.plot(t,f)
plt.xlabel('time'); plt.ylabel('f(t)'); plt.grid()
plt.xlim([0, 30])
# Fourier spectrum
yconst = dt*np.sqrt(2./np.pi)
xconst = 2.*np.pi
fw   = np.fft.fft(f)*yconst
freq = np.fft.fftfreq(t.size, dt)*xconst
# Theoretical Expectation
dw = 2*np.pi/dt/Npts
w = np.arange(0,Npts)*dw
Fwa = alpha/((w-wzero)**2 + alpha**2) + alpha/((w+wzero)**2 + alpha**2)
Fwa = np.sqrt(2./np.pi)/2.*Fwa
# Plot spectrum
plt.subplot(2,1,2)
plt.plot(freq, fw.real,'o',label='FFT results')
plt.plot(w,Fwa,label='Theory')
plt.xlim([-2*wzero, 2*wzero])
plt.legend()
plt.xlabel('w'); plt.ylabel('F(w)'); plt.grid()
plt.tight_layout()
plt.show()
```
